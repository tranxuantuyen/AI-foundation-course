```python
import numpy as np
import pandas as pd
def visualize_iris_feature(feature_name):
    df_sort = df.sort_values(feature_name).reset_index(drop = True)
    species = np.array(df_sort["Species"])
    feature_data = np.array(df_sort[feature_name])
    fig, ax =plt.subplots()
    for i in range(len(species)):
        if species[i] == "versicolor":
            ax.scatter(i, feature_data[i], color = "blue")
        elif species[i] == "virginica":
            ax.scatter(i, feature_data[i], color = "red")
        else:
            ax.scatter(i, feature_data[i], color = "black")
    ax.legend()
    plt.show()
    # YOUR CODE HERE
    
```

```python
visualize_iris_feature("Sepal.Width")
```

```python
visualize_iris_feature("Petal.Width")
```

# Bài tập 1

```python
from scipy import stats
import matplotlib.pyplot as plt 
import numpy as np
X =stats.binom.rvs(50, 0.7, size = 5000, random_state = 123)

print("X min = ", X.min())
print("X max = ", X.max())
print("Q1 = ", np.percentile(X,25))
print("Q2 = ", np.percentile(X,50))
print("Q3 = ", np.percentile(X,75))
plt.boxplot(X)
```

```python
plt.hist(X, bins = 50)
plt.title('Do thi tan so')
plt.ylabel('Tan so')
plt.xlabel('Gia tri')
```

```python
plt.hist(X, bins = 10, density=True)
plt.title("Do thi tan suat")
plt.ylabel('Tan so')
plt.xlabel('Gia tri')
```

```python
import seaborn as sns
kwargs = dict(hist_kws={'alpa': 5}, kde_kws={'linewidth': 2})
sns.distplot(X, color='green', bins = 10, label="Ham mat do")
```

```python
sns.distplot(X, color='green', hist=False, bins = 10, label="ham mat do")
```

# Bài tập 2

```python
mean = [165, 55]
cov = [[100, 0.4],[0.4,36]]
X2, Y2 = np.random.multivariate_normal(mean, cov, 5000).T

print("Chieu cao min = ", X2.min())
print("Chieu cao max = ", X2.max())
print("Q1 = ", np.percentile(X2,25))
print("Q2 = ", np.percentile(X2,50))
print("Q3 = ", np.percentile(X2,75))
plt.boxplot(X2)
```

```python
print("Can nang min = ", Y2.min())
print("Can nang max = ", Y2.max())
print("Q1 = ", np.percentile(Y2,25))
print("Q2 = ", np.percentile(Y2,50))
print("Q3 = ", np.percentile(Y2,75))
plt.boxplot(Y2)
```

```python
plt.hist(X2, bins = 50)
plt.title('Do thi tan so')
plt.ylabel('Tan so')
plt.xlabel('Can nang')
```

```python
plt.hist(Y2, bins = 50)
plt.title('Do thi tan so')
plt.ylabel('Tan so')
plt.xlabel('Chieu cao')
```

# Bai tap 1

```python
import numpy as np
x = np.arange(1,7)
y = np.arange(1,7)
jpmf = 1/36*np.ones((6,6))#

pmfx = []
for i in x:
    pmfx.append(jpmf[:,i-1].sum())

pmfy=[]
for j in y:
    pmfy.append(jpmf[j-1,:].sum())
    
EX = np.sum(x*pmfx)
EY = np.sum(y*pmfy)

varX = np.sum((x-EX)**2*pmfx)
varY = np.sum((x-EY)**2*pmfy)
covXY = EXY-EX*EY

cov_matrix = np.array([[varX, covXY], [covXY, varY]])
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import make_circles
x, y = make_circles(n_samples=100,noise=0.05,factor=0.5)

plt.figure()
plt.scatter(x[:,0],x[:,1])

plt.show()
```

```python
W = np.ones((len(x), len(x)))
for i in range(len(x)):
    for j in range(len(x)):
        W[i][j] = np.exp(-100*np.linalg.norm(x[i] - x[j])**2) 
D = np.diag(W.sum(axis = 1))
L = D - W
d,v=eigh(L)
Vk = v[:,:2]
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=2, random_state=0).fit(Vk)
#kmeans.labels_
    
```

```python
x0 = x[kmeans.labels_ == 0]
x1 = x[kmeans.labels_ == 1]
plt.scatter(x0[:, 0], x0[:, 1], color = 'red')
plt.scatter(x1[:, 0], x1[:, 1], color = 'blue')

```

```python
plt.scatter(x[:, 0], x[:,1], c = kmeans.labels_, cmap = 'rainbow', alpha= 0.7)#, edgecolors='b')
```
